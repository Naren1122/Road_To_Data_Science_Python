```python

inventory = {
    'PROD101': {'name': 'Wireless Mouse', 'stock': 12},
    'PROD102': {'name': 'Keyboard', 'stock': 20},
    'PROD103': {'name': 'USB Cable', 'stock': 0},
    'PROD104': {'name': 'Webcam', 'stock': 8}
}


low_stock_warnings = []
out_of_stock_discontinued = []


for product_id in list(inventory.keys()):
    stock = inventory[product_id]['stock']
    name = inventory[product_id]['name']

 
    if stock == 0:
        out_of_stock_discontinued.append(product_id)
        del inventory[product_id]  

   
    elif stock < 15:
        low_stock_warnings.append(name)


print("Low Stock Warnings:", low_stock_warnings)
print("Discontinued Products:", out_of_stock_discontinued)
print("Updated Inventory:", inventory)
```
