```python
allowed_attempts = 5

while allowed_attempts > 0:
    password = input("Enter your password: ")

  
    if len(password) < 8:
        print("Rejected: Password must be at least 8 characters long.")
        allowed_attempts -= 1

    elif " " in password:
        print("Rejected: Password cannot contain spaces.")
        allowed_attempts -= 1

    else:
        print("Password accepted. Access granted.")
        break

    print("Remaining attempts:", allowed_attempts)

# Security lockout if attempts reach 0
if allowed_attempts == 0:
    print("SECURITY LOCKOUT: Too many failed attempts!")
```
