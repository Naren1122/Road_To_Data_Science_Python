```python
def calculate_weekly_stipend(base_hourly_pay, hours_worked, overtime_multiplier=1.5):
    standard_hours = 40

   
    if hours_worked <= standard_hours:
        standard_pay = hours_worked * base_hourly_pay
        overtime_pay = 0
    else:
        standard_pay = standard_hours * base_hourly_pay
        overtime_hours = hours_worked - standard_hours
        overtime_pay = overtime_hours * base_hourly_pay * overtime_multiplier

   
    total_gross_pay = standard_pay + overtime_pay

    return total_gross_pay


pay1 = calculate_weekly_stipend(20, 38)   
pay2 = calculate_weekly_stipend(20, 45)   

print("Employee 1 Weekly Pay: $", pay1)
print("Employee 2 Weekly Pay: $", pay2)
```
