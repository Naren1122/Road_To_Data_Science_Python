```python
def verify_badge(badge_dict, badge_id):
    if badge_id in badge_dict:
        return badge_dict[badge_id]   # Return employee name
    else:
        return "Access Denied: Invalid Badge ID"

employee_db = {
    "ID101": "Shyam",
    "ID102": "Sita"
}

# Example usage
print(verify_badge(employee_db, "ID101"))  
print(verify_badge(employee_db, "ID999"))  
```
