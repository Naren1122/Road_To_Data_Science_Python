```python
def calculate_average_score(total_points, total_ratings):
    try:
        average = total_points / total_ratings
        return average
    except ZeroDivisionError:
        return 0.0

score1 = calculate_average_score(85, 17)   
score2 = calculate_average_score(50, 0)    

print("Average Score (Workshop 1):", score1)
print("Average Score (Workshop 2):", score2)
```
