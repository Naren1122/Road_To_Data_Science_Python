```python

durations = [45, 120, 310, 15, 640, 180]


price_tiers = {
    "$5 Tier": 0,
    "$12 Tier": 0,
    "$30 Tier": 0
}


total_revenue = 0


for duration in durations:
    if duration < 60:
        price_tiers["$5 Tier"] += 1
        total_revenue += 5

    elif duration <= 240:
        price_tiers["$12 Tier"] += 1
        total_revenue += 12

    else:
        price_tiers["$30 Tier"] += 1
        total_revenue += 30


print("Vehicle Count by Price Tier:", price_tiers)
print("Total Revenue: $", total_revenue)
```
