```python
cart_total = float(input("Enter cart total ($): "))
region = input("Enter delivery zone (Domestic/International): ").stip().lower()
premium_member = input("Are you a premium member? (yes/no): ").strip().lower()
```

```python
if cart_total > 150:
    shipping_cost = 0
else:
    if region == "domestic":
        shipping_cost = 10
    elif region == "international":
        shipping_cost = 25
    else:
        print("Invalid delivery zone!")
        exit()

    if premium_member == "yes":
        shipping_cost -= 5

    if shipping_cost < 0:
        shipping_cost = 0
```

```python
print(f"Final shipping cost: ${shipping_cost}")
```
