```python
import numpy as np
arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape(2,3)
print(reshaped)
```

```python
reshaped = arr.reshape(3,2)
print(reshaped)
```

```python
#conbert to 1d
```

```python
ravel = reshaped.ravel()
ravel
```

```python
#flatten to convert to 1D which returns the copy
flat = reshaped.flatten()
flat
```
