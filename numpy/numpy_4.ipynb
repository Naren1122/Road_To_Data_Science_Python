```python
# matrix_operations .....................................
import numpy as np
a = np.array([[1,2], [3,4]])
b = np.array([[5,6],[7,8]])

a @ b # this is the multiplication on the matrix using @
```

```python
# stacking array --> there are3 types of stack ie vertical , horizontal and column_stack

a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])

np.vstack((a,b ))# vertical stack
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b)) # column stack
```
