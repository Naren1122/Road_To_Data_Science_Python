```python
import numpy as np
a1 = np.array([1,2,3,4,5]) #the size of both array should same else it will throw error.
a2 = np.array([6,7,8,9,10])
result = a1+a2
print (result)


#similaly u can perform other arithematic operation --> + - * % ** // etc
```

```python
#Broadcasting ...............................................

a1 = [1,2,3,4,5,6,7]
arr = np.array(a1)
result = arr + 10 #brocasting mean if u have to perform operation in all the element of arr of different shape
print (result)
```

```python
#deep and shallow copy

'''here we assigned the value a = b , and changed the value of 0 index using b then it
changed a also so this is shallow copy'''
a = [1,2,3,4,5]
b = a
b[0] = 44
b 
```

```python
'''here if i copy the value of a to b using copy then when change of b is made it does not affect a 
this is deep copy'''

b = a.copy()
b[0] =90
b
```
