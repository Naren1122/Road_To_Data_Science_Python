```python
'''
Abstraction --->Data hiding and what is neccessary[mobile banking]

rule --> need to create abstract class
         atleast one function must be an abstract function 

'''

#abstract base class
from abc import ABC, abstractmethod

class BankAccount(ABC):   # Abstract class

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def deposit(self, amount):
        pass
```

```python
class SavingsAccount(BankAccount):
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}")
```

```python
account = SavingsAccount(1000)

account.deposit(500)
account.withdraw(300)
print(account.balance)
```
