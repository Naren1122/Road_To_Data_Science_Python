```python
from abc import ABC, abstractmethod
class Appliance(ABC):
    
    @abstractmethod
    def start_work(self):
        pass
```

```python
class WashingMachine(Appliance):
    def start_work(self):
        print("Drum is spinning at 1200rpm")
```

```python
class Dishwasher(Appliance):
    def start_work(self):
        print("High-pressure water jets activated....")
```

```python
wm = WashingMachine()
dw = Dishwasher()

wm.start_work()
dw.start_work()
```
