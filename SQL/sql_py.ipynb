```python
import sqlite3
```

```python
conn = sqlite3.connect('sql.db')
```

```python
cursor = conn.cursor()
```

```python
try:
    cursor.execute('''
    Create Table Customers(
    customer_id integer primary key autoincrement,
    customer_name text,
    customer_address text,
    phone text,
    -- premium,normal,vip
    customer_type text);
    ''')
    
    print('Customer created successfully.')
except:
    print('Warning : Customer table already created.')    
```

```python
try:
    cursor.execute(
    '''
    Create Table Staffs(
    staff_id integer primary key autoincrement,
    staff_name text,
    staff_phone text,
    --senior,associate, junior
    job_position text,
    --Full-Time,Part-Time,Contract
    job_type text
    );
    '''
    )
    print('Staff created successfully.')
except:
    print('Warning : Staff Table already created.')
```

```python
try:
    cursor.execute(
    '''
    create table orders(
    order_id integer primary key autoincrement,
    order_date date,
    --completed, pending,processing, rejected
    order_status text,
    order_quantity integer,
    order_price real,
    order_destination text,
    customer_id integer,
    staff_id integer,
    foreign key (customer_id) references customers (customer_id),
    foreign key(staff_id) references staffs (staff_id)
    )    
    '''
    )
    print('Orders created Successfully')

except:
    print('Warning : Order table already created.')
```
