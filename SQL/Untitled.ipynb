```python
'''
sqlite datatypes

- Integer
- Text
- Real
- Numeric
- BLOB -> Binary Data
- CLOB -> Character Large Object
-any


Contraints -> Rules of tables

- Primary Key -> Unique id which define a single row
- Foreign Key -> Used to create relationship between two tables
- Null -> Can insert null values
- Not Null -> Must insert values 
- Check -> to check validation of data
- Unique -> only unique data can be inserted

Relation

- one to one
- many to one
- one to many
- many to many

Cardinality

- Mandatory -> |
- Optional -> o
'''
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('sql.db')
```

```python
#query communication between python and sqlite 
cursor = conn.cursor()
```

```python
try:
    cursor.execute('''
    Create Table Customers(
    customer_id integer primary key autoincrement,
    customer_name text,
    phone text,
    -- premium,normal,vip
    customer_type text);
    ''')
    
    print('Customer created successfully.')
except:
    print('Customer table already created.')    
```

```python
try:
    cursor.execute(
    '''
    Create Table Staffs(
    staff_id integer primary key autoincrement,
    staff_name text,
    staff_phone text,
    --senior,associate, junior
    job_position text,
    --Full-Time,Part-Time,Contract
    job_type text
    );
    '''
    )
    print('Staff created successfully.')
except:
    print('Staff Table already created.')
```

```python
try:
    cursor.execute(
    '''
    create table orders(
    order_id integer primary key autoincrement,
    order_date date,
    --completed, pending,processing, rejected
    order_status text,
    order_quantity integer,
    order_price real,
    order_destination text,
    customer_id integer,
    foreign key (customer_id) references customers (customer_id),
    foreign key(staff_id) references staffs (staff_id)
    )    
    '''
    )
    print('Orders created Successfully')

except:
    print('Order table already created.')
```
