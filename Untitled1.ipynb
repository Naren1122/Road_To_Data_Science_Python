```python
class AmountDepositError(Exception):
    pass
class AmountWithdrawError(Exception):
    pass
```

```python
import random
class Bank:
    def __init__(self,account_name,initial_balance):
        self.account_name = account_name
        self.initial_balance = initial_balance
        self.account_number = "".join(str(random.randint(0,9)) for i in range(16))
        
    def deposit(self,amount):
        if amount > 100:
            self.initial_balance += amount
            print(f'Rs. {amount} has been deposited to A/C no. {self.account_number}.')
        else:
            raise AmountDepositError('Error: Amount must be more than 100.')
            
    def withdraw(self,amount):
        if amount < self.initial_balance:
            self.inital_balance -= amount
            print(f'{amount} has been withdrawn from the A/C no. {self.account_number}')
        else:
            raise AmountWithdrawError ('Error: Withdrawn amount must be less than balance amount')
            
        
        
    def show_details(self):
        print(self.account_name)
        print(self.initial_balance)
        print(self.account_number)
```
