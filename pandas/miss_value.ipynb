```python
import numpy as np
import pandas as pd
```

```python
employees = {
    "name": ["Alice", "Bob", np.nan, "David"],
    "age": [25, np.nan, 28, 35],
    "city": ["New York", "Chicago", np.nan, "Seattle"],
    "salary": [75000, 82000, np.nan, 95000]
}
```

```python
df = pd.DataFrame(employees)
df
```

```python
df.isna() # find if there is null value or not which is np.nan
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#REMOVE MISSING VALUE...............................................
```

```python
df.dropna() # remove null value
```

```python
df.dropna(thresh = 4) # in column if there is 3 not null value then (non_null) then tyo not null row bahek sab row hatxa
```

```python
# FILL ALL THE MISSING DATA .................................................
```

```python
df.fillna(0)
```

```python
values = {"name":"Ravan","age":34,"city":"Kathmandu","salary":23456}
df.fillna(value = values)
```
