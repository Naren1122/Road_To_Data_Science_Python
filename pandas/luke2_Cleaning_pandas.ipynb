```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('csv', data_files='data_jobs.csv')
```

```python
# 1. Convert the column to datetime safely
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')

# 2. Extract the month
df['job_posted_month'] = df['job_posted_date'].dt.month

# 3. Look at the first 5 rows to make sure it worked
df[['job_posted_date', 'job_posted_month']].head() #converting str to datetime using t0_datetime() function method creating job_posted_date column 
```
