```python
import numpy as np
import pandas as pd
```

```python
customers = {
    "customer_id": [201, 202, 203, 204],
    "name": ["John", "Emma", "Noah", "Olivia"],
    "city": ["Kathmandu", "Pokhara", "Lalitpur", "Biratnagar"]
}

customers = pd.DataFrame(customers)
customers
```

```python
orders = {
    "order_id": [1001, 1002, 1003, 1004, 1005],
    "customer_id": [201, 202, 202, 205, 203],
    "amount": [500, 700, 300, 1000, 450]
}

orders = pd.DataFrame(orders)
orders
```

```python
pd.merge(customers,orders)
```

```python
pd.merge(customers,orders,on='customer_id',how='outer')
```

```python
pd.merge(customers,orders,on='customer_id',how='inner')
```

```python
pd.merge(customers,orders,on='customer_id',how='left')
```

```python
#Concatination of 2 dataframe --------------------------------------------------
```

```python
pd.concat([customers,orders])
```

```python
pd.concat([customers,orders],axis = 1)
```
