```python
# working on dictionary
fruit_protein = {
    "Apple": 0.3,
    "Banana": 1.1,
    "Orange": 0.9,
    "Mango": 0.8,
    "Papaya": 0.5,
    "Pineapple": 0.5,
    "Guava": 2.6,
    "Kiwi": 1.1,
    "Watermelon": 0.6,
    "Grapes": 0.7,
    "Strawberry": 0.8,
    "Blueberry": 0.7,
    "Pear": 0.4,
    "Peach": 0.9,
    "Plum": 0.7,
    "Cherry": 1.1,
    "Pomegranate": 1.7,
    "Avocado": 2.0,
    "Dragon Fruit": 1.2,
    "Jackfruit": 1.7
}

s1 = pd.Series( fruit_protein,name="Protein")
print(s1.to_frame())
```

```python
#Conditional Selection.......................................................................................................
s1[s1>1]
```

```python
#logical operator and or not...............................................................................................................
s1[(s1>0.5) & (s1<1)]

```
