```python
import numpy as np
import pandas as pd
```

```python
employees = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace"],
    "department": ["HR", "IT", "HR", "IT", "Finance", "Finance", "IT"],
    "salary": [50000, 70000, 55000, 80000, 60000, 65000, 72000]
}
df = pd.DataFrame(employees)
df
```

```python
r = df.groupby('department')['salary'].sum()
r
```

```python
df["category"] = ['A','B','A','C','D','F','G']
df
```

```python
r = df.groupby(['category','department'])['salary'].sum()
r
```

```python
mean =df['salary'].mean() # min max count std
print(mean)
```
