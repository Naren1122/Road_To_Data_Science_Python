```python
"""
try-except
else
finally

custom exception

if we use if ...else the error will be shown in terminal only but not to the user...
"""
try:
    num1 = int(input("Enter a num1: "))
    num2 = int(input("Enter a num2: "))
    result = num1/num2
    print(f"{result}")
except ZeroDivisionError:
    print("Error: Number cannot be divisible by 0")
except ValueError:
    print('Error: only inter values are accepted in the input field(0-9).')
```

```python
try:
    file_name = input("Enter a file name:")
    with open("file_name","r") as file:
        data = file.read()
        print(data)
except FileNotFoundError:
    print(f"Error: No such file found in folder named: {file_name}.txt")
else:
    print(data)
finally:
    print('Program ended successfully.')
```

```python
#custom exception
class AmountWithDrawError(Exception):
    pass
```

```python
balance =5000
def withdrawAmount():
    global balance
    try:
        withdraw_amount = float(input("Enter withdraw amount: "))
    except ValueError:
        print("Invalid input! Please enter a numeric value.")
        return 
    
    if withdraw_amount <= 0:
        print("Amount must be greater than 0.")
        return

    if withdraw_amount <= balance:
        balance -= withdraw_amount
        print(f"Rs.{withdraw_amount} successfully withdrawn from account.")
        print(f"Remaining balance: Rs.{balance}")
    else:
        raise AmountWithDrawError("Amount must be less than or equal to actual balance.")
   
```

```python

try:
    withdrawAmount()
except AmountWithDrawError as e:
    print(e)
```
