```python
from selenium import webdriver # access browser and website
from selenium.webdriver.common.by import By # access web elements (tags,css, xpath)
```

```python
website = "https://merolagani.com/LatestMarket.aspx"
```

```python
driver = webdriver.Chrome() # used for edge browser so 
# driver = webdriver.chrome() # user for chrome
```

```python
driver.get(website)
```

```python
table_body = driver.find_element(By.TAG_NAME,'tbody')
```

```python
table_rows = table_body.find_elements(By.TAG_NAME,'tr')
```

```python
# for data in table_rows:
#     columns = data.find_elements(By.TAG_NAME,'td')
# for col in columns:
#     print(col.text)
```

```python
symbol_col= []
ltp_col = []
change_col =[]
open_col=[]
high_col = []
low_col = []
qty_col = []
pclose_col = []
diff_col = []
```

```python
for row in table_rows:
    columns = row.find_elements(By.TAG_NAME, 'td')

    if len(columns) < 9:
        continue
    symbol_col.append(columns[0].text)
    ltp_col.append(columns[1].text)
    change_col.append(columns[2].text)
    open_col.append(columns[3].text)
    high_col.append(columns[4].text)
    low_col.append(columns[5].text)
    qty_col.append(columns[6].text)
    pclose_col.append(columns[7].text)
    diff_col.append(columns[8].text)
```

```python
symbol_col
```

```python
data_frame = {
    'Symbol': symbol_col,
    'LTP': ltp_col,
    'Change': change_col,
    'Open': open_col,
    'High': high_col,
    'Low': low_col,
    'Quantity': qty_col,
    'P_Close': pclose_col,
    'Difference': diff_col
}
```

```python
driver.quit()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data_frame)
```

```python
df.to_csv('stocks.csv', index=False)
# data.to_json('stocks.csv', index=False)
```
