```python
#next question
'''
type of variable:
-global : outside the block scope,
local : variable defined inside constructor, function,
instant
'''

class Employees:
    def __init__(self, employee_name, employee_position, employee_salary):
        self.employee_name = employee_name
        self.employee_position = employee_position
        self.employee_salary = employee_salary

    # for oop
    def employee_details(self):
        print(self.employee_name)
        print(self.employee_position)
        print(self.employee_salary)
```

```python
all_employees = []
```

```python
for i in range(2):
    employee_name = input("Enter your name: ")
    employee_position = input("Enter your position: ")
    employee_salary = int(input("Enter your salary: "))

    e = Employees(employee_name, employee_position, employee_salary)

    
    all_employees.append(e)


print("\nEmployee Details:")
for emp in all_employees:
    emp.employee_details()
```
