```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.allrecipes.com/search?q=chicken")
```

```python
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "mntl-search-results__list_1-0"))
)
```

```python
recipe = driver.find_elements(By.CSS_SELECTOR, ".mntl-card-list-card--extendable")
```

```python
results =[]
```

```python
for card in recipe:
      
        title = card.find_element(By.CSS_SELECTOR, ".card__title-text").text.strip()
        
        rating = card.find_element(By.CSS_SELECTOR, "[class*='rating-count-number']").text.strip()
           
        rating = rating.split()[0]
        

        
        results.append({
            "Recipe Title": title,
            "Ratings Count": rating,
           
        })
        
```

```python
driver.quit()
```

```python
df = pd.DataFrame(results)

print("\n" + "═"*90)
print("                           ALLRECIPES SCRAPER DATA REPORT                           ")
print("═"*90)
print(df.to_markdown(index=False, tablefmt="grid"))
print("═"*90)

# 7. Save output to disk
df.to_csv("allrecipes_data.csv", index=False)
print("[SUCCESS] Data cleanly extracted and saved to 'allrecipes_data.csv'")
```
