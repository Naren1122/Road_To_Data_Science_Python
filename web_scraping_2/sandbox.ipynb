```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://sandbox.oxylabs.io/products")
```

```python
WebDriverWait(driver,10).until( #Keep checking the website DOM for up to 10 seconds
    EC.presence_of_all_elements_located((By.CSS_SELECTOR,".product-card"))
)
```

```python
results = []
```

```python
cards = driver.find_elements(By.CSS_SELECTOR,".product-card")
```

```python
for card in cards:
    title = card.find_element(By.CSS_SELECTOR,".title").text.strip()
    price =card.find_element(By.CSS_SELECTOR,".price-wrapper").text.strip()
    results.append(
        {
            "title": title,
            "price":price,
        }
    )
    print(results)
```

```python
time.sleep(2)
```

```python
driver.quit()
```

```python
df = pd.DataFrame(results)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

```

```python
print("\n" + "="*60)
print("                  OXYLABS SANDBOX PRODUCTS                  ")
print("="*60)
print(df.to_string(index=False)) 
print("="*60)

csv_filename = "oxylabs_products.csv"
df.to_csv(csv_filename, index=False)

```
