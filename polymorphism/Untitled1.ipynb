```python
class VideoPlayer:
    def play_content(self):
        return "Streaming video at 1080p"


class AudioPlayer:
    def play_content(self):
        return "Playing high-fidelity audio"


class GIFPlayer:
    def play_content(self):
        return "Looping silent animation"

        
```

```python
def press_play(media_obj):
    print(media_obj.play_content())
```

```python
video = VideoPlayer()
audio = AudioPlayer()
gif = GIFPlayer()

```

```python
press_play(video)
press_play(audio)
press_play(gif)
```
