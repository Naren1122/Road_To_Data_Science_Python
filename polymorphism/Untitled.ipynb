```python
'''
polymorphism
one function multiple similar work
'''
class viber:
    def message(self,send_from,send_to,message):
        print(f"{send_to}")
        print(f"{send_from} sent u a message")
        print(f"Message:{message}")
    
class messenger:
    def message(self,send_from,send_to,message):
        print(f"{send_to}")
        print(f"{send_from} sent u a message")
        print(f"Message:{message}")
    
class whatsapp:
    def message(self,send_from,send_to,message):
        print(f"{send_to}")
        print(f"{send_from} sent u a message")
        print(f"Message:{message}")
    
class slack:
    def message(self,send_from,send_to,message):
        print(f"{send_to}")
        print(f"{send_from} sent u a message")
        print(f"Message:{message}")    
    
```

```python
def send_message(message_obj,send_from,send_to,message):
    message_obj.message(send_from,send_to,message)
```

```python
send_message(whatsapp(),'bob','Shyam','lol')
```

```python
# another method without defining a function send_message
whatsapp().message('bob','shyam','lol')
```
