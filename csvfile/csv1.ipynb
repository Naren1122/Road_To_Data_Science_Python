```python
''' 
A company wants to develop an Employee Attendance Management System using a CSV file named attendance.csv.

Write a Python program that performs the following tasks:

Step 1: Create the CSV File (Write Mode)
Ask the user how many employee records they want to enter.
For each employee, take:
Employee ID
Employee Name
Department
Attendance Status (Present/Absent)
Store all records in attendance.csv.
The first row should contain the column headers.

Example:

Employee_ID	Name	Department	Status
101	John	IT	Present
102	Alice	HR	Absent
Step 2: Read the CSV File (Read Mode)

After creating the file, read all the records from attendance.csv and display them in a formatted manner.

Example Output

Employee_ID    Name      Department    Status
------------------------------------------------
101            John      IT            Present
102            Alice     HR            Absent
Step 3: Append a New Employee (Append Mode)

Now ask the user to enter one more employee's details:

Employee ID
Name
Department
Status

Append this new record to the existing CSV file.

Step 4: Read the Updated File

Finally, display all records again after appending the new employee.

'''
```

```python
'''A company wants to develop an Employee Attendance Management System using a CSV file named attendance.csv.'''

import csv
n = int(input("Enter number of employees:"))

with open("attendance.csv","w",newline="") as file:
    writer = csv.writer(file)
    #write header
    writer.writerow(["Employee_ID","Name","Department","Status"])

    for i in range(5):
        print(f"Enter the details of a users:{i+1}")
        employee_id = int(input("Enter the id"))
        name = input("Enter a name:")
        department= input("Enter department")
        status = input("Enter Status(present/absent):")

        writer.writerow([employee_id,name,department,status])

        print("\n Attendance file created Successfully")
```

```python
#read csv

print("\n=================Employee Attendance======================")
with open("attendance.csv","r",newline="") as file:
    reader = csv.reader(file)
    for r in reader: 
        print("{:<15} {:<15} {:<15} {:<15}".format(*r))
    
   
```

```python
#append csv

print("\n============append csv======================")
employee_id = int(input("Enter a employee_id:"))
name = input("Enter a name:")
department=input("Enter a department:")
status= input("Enter a status (Present/Absent):")

with open("attendance.csv","a",newline="") as file:
    writer = csv.writer(file)

    writer.writerow([employee_id,name,department,status])
    print("\n New employee added successfully!")

    
```

```python
#read updated file

print("\n============read updated file=============")
with open("attendance.csv","r",newline="") as file:
    reader = csv.reader(file)
    for r in reader:
        print("{:<15} {:<15} {:<15} {:<15}".format(*r))
    
```
