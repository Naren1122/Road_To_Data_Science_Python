```python
'''
A supermarket wants to develop a Product Inventory Management System using a CSV file named inventory.csv.

Write a Python program that performs the following tasks:

Step 1: Create the CSV File (Write Mode)

Ask the user how many products they want to enter.

For each product, take the following details:

Product ID
Product Name
Category
Price
Stock Quantity

Store all the records in inventory.csv.

The first row should contain the column headers.

Example
Product_ID	Product_Name	Category	Price	Stock
101	Laptop	Electronics	85000	10
102	Mouse	Electronics	1200	50
Step 2: Read the CSV File (Read Mode)

Read the file and display all the products in a well-formatted table.

Example:

Product_ID   Product_Name   Category      Price    Stock
----------------------------------------------------------
101          Laptop         Electronics   85000    10
102          Mouse          Electronics   1200     50
Step 3: Append a New Product (Append Mode)

Ask the user to enter details of one new product:

Product ID
Product Name
Category
Price
Stock Quantity

Append this product to the existing CSV file.

Step 4: Read the Updated File

Display all the products after appending the new record.

Step 5: Search a Product

Ask the user to enter a Product ID.

If the product exists, display all its details.
Otherwise, print:
Product Not Found
Example
User enters
Enter Product ID to search:
102
Output
Product Found

Product ID   : 102
Product Name : Mouse
Category     : Electronics
Price        : 1200
Stock         : 50
'''
```

```python
import csv

# ---------------------------------
# Step 1: Create CSV File (Write)
# ---------------------------------

n = int(input("Enter the number of products: "))

with open("inventory.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Write Header
    writer.writerow(["Product_ID", "Product_Name", "Category", "Price", "Stock"])

    # Take product details
    for i in range(n):
        print(f"\nEnter details of Product {i+1}")

        product_id = input("Enter Product ID: ")
        product_name = input("Enter Product Name: ")
        category = input("Enter Category: ")
        price = input("Enter Price: ")
        stock = input("Enter Stock Quantity: ")

        writer.writerow([product_id, product_name, category, price, stock])

print("\nInventory file created successfully!")
```

```python

# ---------------------------------
# Step 2: Read CSV File
# ---------------------------------

print("\n========== Product Inventory ==========\n")

with open("inventory.csv", "r", newline="") as file:
    reader = csv.reader(file)

    for row in reader:
        print("{:<12} {:<20} {:<15} {:<10} {:<10}".format(*row))
```

```python

# ---------------------------------
# Step 3: Append New Product
# ---------------------------------

print("\nEnter details of a new product:\n")

product_id = input("Enter Product ID: ")
product_name = input("Enter Product Name: ")
category = input("Enter Category: ")
price = input("Enter Price: ")
stock = input("Enter Stock Quantity: ")

with open("inventory.csv", "a", newline="") as file:
    writer = csv.writer(file)

    writer.writerow([product_id, product_name, category, price, stock])

print("\nNew product added successfully!")
```

```python
# ---------------------------------
# Step 4: Read Updated File
# ---------------------------------

print("\n========== Updated Product Inventory ==========\n")

with open("inventory.csv", "r", newline="") as file:
    reader = csv.reader(file)

    for row in reader:
        print("{:<12} {:<20} {:<15} {:<10} {:<10}".format(*row))
```
