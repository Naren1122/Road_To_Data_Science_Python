```python
#wap that asks the user to enter 5 product name and appends them to a file named products.txt one product name per line. 
# then read the file and display the product names in uppercase.

products = []

for i in range(5):
    product_name = input("Enter product name: ")
    products.append(product_name)

with open("products.txt", "w") as file:
    file.write("\n".join(products))

print("Products saved successfully.")
```
