```python
'''Write a Python program that asks the user for:

Name
Age
Address

Store the information as a dictionary and save it into a JSON file named person.json.'''

import json
name = input("Enter your name: ")
age = int(input("Enter your age: "))
address = input("Enter your address: ")

info ={
    "Name" : name,
    "Age" : age,
    "Address" : address
}
with open("person.json","w") as file:
    json.dump(info,file,indent=4)
    
    for key,values in info.items():
        print(f"{key}:{values}")
   
```

```python
'''Write a Python program that:

Reads a JSON file named student.json.
Adds a new student record to the existing data.
Saves the updated data back into the same file. '''



import json

# Read the existing data
with open("student.json", "r") as file:
    students = json.load(file)

# Take input for a new student
name = input("Enter student name: ")
age = int(input("Enter age: "))
faculty = input("Enter faculty: ")

new_std = {
    "Name": name,
    "Age": age,
    "Faculty": faculty
}

# Add the new record
students.append(new_std)

# Save the updated data
with open("student.json", "w") as file:
    json.dump(students, file, indent=4)

# Display all students
for student in students:
    print(student)
```
