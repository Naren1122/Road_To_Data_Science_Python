```python
'''inheritance
- simple inheritance ---> two classes [parent and child class]
- multiple inheritance ---> 3or more than 3 class ie 1 child and multiple parents
- multi-level inheritance ---> 3 or more than3 classes ie grand parent, parent and child
'''


#simple inheritance..............................................................
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print("\nCustomer Details")
        print(f"Customer ID       : {self.customer_id}")
        print(f"Customer Name     : {self.customer_name}")
        print(f"Customer Address  : {self.customer_address}")
        print(f"Customer Phone    : {self.customer_phone}")
        
        
```

```python
class Product(Customer):
    def __init__(self,product_id,product_name,product_quantity,list_price,product_description,customer_id,customer_name,customer_address,customer_phone):

          #super key is used to inherit the parent class with child class........................ 
        super().__init__(customer_id,customer_name,customer_address,customer_phone)

        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print("\nProduct Details")
        print(f"Product_id: {self.product_id}")
        print(f"Product_name: {self.product_name}")
        print(f"Product_quantity: {self.product_quantity}")
        print(f"List_price: {self.list_price}")
        print(f"Product_description: {self.product_description}")
        print(f"Total_price: {self.total_price}")
```

```python
p = Product(1,"Biscuit",2,200,"tasty biscuit",123,"Naren","Dhapakhel",90000000)
```

```python
p.product_details()
```
