```python
class SavingsAccount:
    def __init__(self,curr_balance):
        self.__balance = curr_balance
        
    def withdraw(self, amount):
        if self.__balance >=500:
            self.__balance -= amount
        else:
            print("Transaction DEnied")
    def get_balance(self):
        return self.__balance        
 
   
    
a = SavingsAccount(1000)
a.withdraw(200)
print(f"Current Balance is {a.get_balance()}")
a.withdraw(2000)
```

```python
class Credential:
    def __init__(self, api_key):
        self.__api_key = api_key  # Private attribute

    def get_masked_key(self):
        # Handle cases where the key is too short to mask effectively
        if len(self.__api_key) <= 6:
            return "***" 
            
        first_three = self.__api_key[:3]
        last_three = self.__api_key[-3:]
        
        # Calculate matching number of asterisks for the middle section
        asterisks_count = len(self.__api_key) - 6
        mask = "*" * asterisks_count
        
        return f"{first_three}{mask}{last_three}"

# --- Verification ---
creds = Credential("ABC123456789XYZ")
print(creds.get_masked_key())
```
