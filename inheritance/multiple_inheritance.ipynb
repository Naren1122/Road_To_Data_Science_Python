```python
#multiple inheritance

class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print("\nCustomer Details")
        print(f"Customer ID       : {self.customer_id}")
        print(f"Customer Name     : {self.customer_name}")
        print(f"Customer Address  : {self.customer_address}")
        print(f"Customer Phone    : {self.customer_phone}")
        
        
```

```python
class Staff:
    def __init__(self,staff_id,staff_name,staff_position,staff_job_type):
        self.staff_id = staff_id
        self.staff_name = staff_name
        self.staff_position = staff_position
        self.staff_job_type = staff_job_type

    def staff_details(self):
        print("\n Staff Details")
        print(f"Staff_Id : {self.staff_id}")
        print(f"Staff_name :{self.staff_name}")
        print(f"Staff_Position : {self.staff_position}")
        print(f"Staff_job_type: {self.staff_job_type}")
```

```python
class Product(Customer, Staff):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone,
                staff_id, staff_name, staff_position, staff_job_type):
        # Constructor inherit and since it works but in production we use super key because it prevents from data leakage of the class
        Customer.__init__(self, customer_id, customer_name, customer_address, customer_phone)
        Staff.__init__(self, staff_id, staff_name, staff_position, staff_job_type)
        
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
#Next Way using **kwargs and super key
# Parent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone, **kwargs):
        super().__init__(**kwargs)
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
class Staff:
    def __init__(self, staff_id, staff_name, staff_position, staff_job_type, **kwargs):
        super().__init__(**kwargs)
        self.staff_id = staff_id
        self.staff_name = staff_name
        self.staff_position = staff_position
        self.staff_job_type = staff_job_type

    def staff_details(self):
        print(f'Staff Id: {self.staff_id}')
        print(f'Staff Name: {self.staff_name}')
        print(f'Staff Position: {self.staff_position}')
        print(f'Staff Job Type: {self.staff_job_type}')
```

```python
class Product(Customer, Staff):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone,
                staff_id, staff_name, staff_position, staff_job_type):
        # Constructor Inheritance
        super().__init__(customer_id=customer_id, customer_name=customer_name, customer_address=customer_address, 
                         customer_phone=customer_phone, 
                         staff_id=staff_id, staff_name=staff_name, 
                        staff_position=staff_position, staff_job_type=staff_job_type)
        
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python

class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print("\nCustomer Details")
        print(f"Customer ID       : {self.customer_id}")
        print(f"Customer Name     : {self.customer_name}")
        print(f"Customer Address  : {self.customer_address}")
        print(f"Customer Phone    : {self.customer_phone}")
        
```

```python
class Product(Customer):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone
                ):
        # Constructor inherit and since it works but in production we use super key because it prevents from data leakage of the class
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
      
        
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
class Order(Product):
    def __init__(self, order_id, order_desc, order_destination, order_quantity, product_id, product_name, product_quantity, list_price, product_description,
                 customer_id, customer_name, customer_address, customer_phone):
        super().__init__(product_id, product_name, product_quantity, list_price, product_description,
                         customer_id, customer_name, customer_address, customer_phone)
        
        self.order_id = order_id
        self.order_desc = order_desc
        self.order_destination = order_destination
        self.order_quantity = order_quantity
        
    def order_details(self):
        print('Order_Details')
        print('-'*55)
        print(f'Order ID:{self.order_id}')
        print(f'Order Desc:{self.order_desc}')
        print(f'order_destination: {self.order_destination}')
        print(f'order_quantity:{self.order_quantity}')
```
