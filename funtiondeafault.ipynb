```python
#using all arguments

def product_info(*product_id, product_name, brand_id, category_id, model_year, last_price=0):
    print(f"product_id: {product_id}")
    print(f"product_name: {product_name}")
    print(f"brand_id: {brand_id}")
    print(f"category_id: {category_id}")
    print(f"model_year: {model_year}")
    print(f"last_price: {last_price}")


# Function call
product_info(
    1, 2, 3, 4, 5,
    product_name="iPhone",
    brand_id=2,
    category_id=3,
    model_year=2020
)
```

```python
def products(**pros):
    for key, value in pros.items():
        print(f"{key}: {value}")


products(
    product_id=1,
    product_name="Laptop",
    brand_id=1,
    category_id=10,
    model_year=2024,
    last_price=999.99
)
```
